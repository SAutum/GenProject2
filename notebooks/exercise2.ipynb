```python
import torch
from torch.utils.data.dataloader import DataLoader

from data_utils.BrainDataset import BrainDataset
from data_utils.BrainSampler import BrainSampler
from data_utils import visualize_batch
```

```python
torch.random.manual_seed(1)
training_brains = {(1818, 759, "left"), (1818,820, "right"), (1818, 830, "right")}
map_type = "transmittance"
data = BrainDataset(training_brains, map_type, resolution="05")
sampler = BrainSampler(data, tile_size=64, map_type=map_type)
loader = DataLoader(data, sampler=sampler, batch_size=8)
batch = next(iter(loader))[0]
visualize_batch(batch, map_type)
```
